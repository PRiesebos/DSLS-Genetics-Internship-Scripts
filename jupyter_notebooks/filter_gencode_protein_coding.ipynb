```python
import pandas as pd
```

```python
gen = pd.read_csv("/scratch/hb-functionalgenomics/projects/gut-bulk/reference/gencode_44_2023/gencode.v44.primary_assembly.annotation.gtf", sep="\t", skiprows=5, names=["chromosome","annotation_source","feature_type","genomic_start_location","genomic_end_location","score","genomic_strand","genomic_phase","key_value_pairs"])
gen
```

```python
# Filter for protein-coding genes
protein_coding_genes = gen[gen['key_value_pairs'].str.contains('gene_type "protein_coding";') & (gen['feature_type'] == "gene")]

# Split the 'key_value_pairs' column by ';' into multiple columns
split_columns = protein_coding_genes['key_value_pairs'].str.split(';', expand=True)

# Assign relevant column names to the first few columns of interest
split_columns.columns = ["gene", "gene_type", "gene_name", "hgnc_id", "havana_gene", *range(6, 11)]

# Further split the 'gene' column and clean up quotes
gene_info = split_columns['gene'].str.split(' ', expand=True).applymap(lambda x: x.replace('"', ''))
gene_info.columns = ["gene_id", "gene_name"]

# Save the 'gene_name' column to a file
# gene_info["gene_name"].to_csv("./protein_coding_genes_list.txt", header=False, index=False)
gene_info
```
