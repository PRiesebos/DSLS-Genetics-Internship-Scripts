# GTEX and mbQTL comparison

```python
# Step 1: Import the 22 chromosome files
file_paths = glob.glob("/scratch/hb-functionalgenomics/projects/gut-bulk/reference/GTEx-V8-eQTLs-b38/Colon_Transverse.v8.EUR.allpairs.chr*.txt.gz")
chromosome_dfs = [pd.read_csv(file, sep="\s+") for file in file_paths]

# Define the mbqtl dataframe
mbqtl = pd.read_csv('/scratch/hb-functionalgenomics/projects/gut-bulk/ongoing/2024-02-07-GutPublicRNASeq/extra_scripts/eQTL_Scripts/count_signif_all.txt', sep="\s+")
mbqtl['SNPPos'] = pd.to_numeric(mbqtl['SNPPos'])
mbqtl['GeneChr'] = pd.to_numeric(mbqtl['GeneChr'])

# Function to process each chromosome dataframe
def process_chr_df(chr_df):
    split_columns = chr_df['variant_id'].str.split('_', expand=True)
    chr_df['alleles'] = split_columns[2] + '/' + split_columns[3]
    chr_df['phenotype_id'] = chr_df['phenotype_id'].str.split('.').str[0]
    chr_df['SNPPos'] = pd.to_numeric(split_columns[1])
    chr_df['chrom'] = pd.to_numeric(split_columns[0].str.replace('chr', ''))
    return chr_df

# Step 2: Apply the processing steps to each dataframe and merge with mbqtl
merged_dfs = []
for i, chr_df in enumerate(chromosome_dfs, start=1):
    print(f"Progress: {i}/22")
    chr_df = process_chr_df(chr_df)
    merged_df = pd.merge(
        mbqtl, chr_df, 
        how='inner', 
        left_on=["Gene", "SNPPos", "GeneChr", "SNPAlleles"], 
        right_on=["phenotype_id", "SNPPos", "chrom", "alleles"]
    )
    merged_dfs.append(merged_df)

# Step 4: Concatenate all merged dataframes into a single dataframe
final_merged_df = pd.concat(merged_dfs, ignore_index=True)
```
